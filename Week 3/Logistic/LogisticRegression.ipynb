```python
import numpy as np
import pandas as pd
from sklearn.metrics import roc_auc_score
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from scipy.spatial import distance

print(0)

data = pd.read_csv('data-logistic.csv', names=['target', 'x1', 'x2'])

X = data[['x1', 'x2']].copy()
y = data['target'].copy()

l = len(data.index) #205

print(1)

w = [0,0]
tmp = [1,1]
k = 0.1
eps = 1e-5
Iter = 0
C = 0

print(2)

while Iter < 1000:
    sum_1 = 0
    sum_2 = 0
    for i in range(l):
        sum_1 += y[i]*X['x1'][i]*(1 - 1/(1 + np.exp(-y[i]*(w[0]*X['x1'][i]+w[1]*X['x2'][i]))))
        sum_2 += y[i]*X['x2'][i]*(1 - 1/(1 + np.exp(-y[i]*(w[0]*X['x1'][i]+w[1]*X['x2'][i]))))
    tmp[0] = w[0]
    tmp[1] = w[1]
    w[0] += k*sum_1/l - k*C*w[0]
    w[1] += k*sum_2/l - k*C*w[1]
    Iter+=1
    if Iter%10==0:
        print(w[0], w[1], sum_1, sum_2, Iter)
    if distance.euclidean(w, tmp) < eps:
        print(w[0], w[1])
        break
        
y_score = np.empty(l)

for i in range(l):
    y_score[i] = 1/(1 + np.exp(-0.1567875*X['x1'][i] - 0.04912*X['x2'][i]))
    
print(roc_auc_score(y, y_score))
    
y_score_ = np.empty(l)

for i in range(l):
    y_score_[i] = 1/(1 + np.exp(-0.028558754546234223*X['x1'][i] - 0.02478013724973556*X['x2'][i]))

print(roc_auc_score(y, y_score_))
```
